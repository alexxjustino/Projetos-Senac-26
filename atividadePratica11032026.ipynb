```python
import pandas as pd
import numpy as np
```

```python
dados = "https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv"

df = pd.read_csv(dados, sep= ";", encoding= "latin1")
```

```python
df_capital = df[df['regiao'] == 'Capital']

veiculos_por_delegacia = df_capital.groupby('cisp')[['roubo_veiculo', 'furto_veiculos']].sum()

veiculos_por_delegacia['total_ocorrencias_veiculos'] = veiculos_por_delegacia['roubo_veiculo'] + veiculos_por_delegacia['furto_veiculos']

veiculos_por_delegacia_sorted = veiculos_por_delegacia.sort_values('total_ocorrencias_veiculos', ascending=False)

veiculos_por_delegacia_sorted
```
