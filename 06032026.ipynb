```python
import pandas as pd
import numpy as np
```

```python
dados_1sem = "Preços semestrais - AUTOMOTIVOS_2025.01.csv"
df_1sem = pd.read_csv(dados_1sem, encoding="latin-1", sep=";")
df_1sem.head(3)
```

```python
dados_2sem = "Preços semestrais - AUTOMOTIVOS_2025.02.csv"
df_2sem = pd.read_csv(dados_2sem, encoding="latin-1", sep=";")
df_2sem.head(3)
```

```python
df_2sem["CNPJ da Revenda"] = df_2sem["CNPJ da Revenda"].astype(str)
```

```python
df_2sem.info()
```

```python
df_anp = pd.concat([df_1sem, df_2sem], ignore_index=True)

```

```python
df_anp["Valor de Venda"] = df_anp["Valor de Venda"].str.replace(",", ".").astype(float)
```

```python
df_gasolina = df_anp.loc[df_anp["Produto"] == "GASOLINA"]
df_etanol = df_anp.loc[df_anp["Produto"] == "ETANOL"]
```

```python
df_media_est_gaso = df_gasolina.groupby("Estado - Sigla")["Valor de Venda"].mean().reset_index()
df_media_est_gaso = df_media_est_gaso.sort_values(by ="Valor de Venda", ascending=False)
df_media_est_gaso
```

```python
df_media_est_eta = df_etanol.groupby("Estado - Sigla")["Valor de Venda"].mean().reset_index()
df_media_est_eta = df_media_est_eta.sort_values(by ="Valor de Venda", ascending=False)
df_media_est_eta
```

```python
# Ranking dos cinco estados com maiores preços para gasolina e etanol
ranking_gasolina = df_media_est_gaso.head(5).reset_index(drop=True)
ranking_etanol = df_media_est_eta.head(5).reset_index(drop=True)

ranking_combustiveis = pd.DataFrame({
    "Estado - Sigla Gasolina": ranking_gasolina["Estado - Sigla"],
    "Preço Gasolina": ranking_gasolina["Valor de Venda"],
    "Estado - Sigla Etanol": ranking_etanol["Estado - Sigla"],
    "Preço Etanol": ranking_etanol["Valor de Venda"]
})

ranking_combustiveis
```

```python
# Análise das concentrações e dispersões regionais nos preços de gasolina e etanol

# Calculando estatísticas de concentração e dispersão por estado
estatisticas_gasolina = df_gasolina.groupby("Estado - Sigla")["Valor de Venda"].agg(['mean', 'std', 'min', 'max', 'count']).reset_index()
estatisticas_etanol = df_etanol.groupby("Estado - Sigla")["Valor de Venda"].agg(['mean', 'std', 'min', 'max', 'count']).reset_index()

# Renomeando colunas para clareza
estatisticas_gasolina.columns = ["Estado", "Preço Médio", "Desvio Padrão", "Mínimo", "Máximo", "Nº de Postos"]
estatisticas_etanol.columns = ["Estado", "Preço Médio", "Desvio Padrão", "Mínimo", "Máximo", "Nº de Postos"]

# Possíveis causas: estados com maior desvio padrão podem indicar influência de logística, demanda local ou competição
analise_regional = pd.DataFrame({
    "Estado": estatisticas_gasolina["Estado"],
    "Gasolina - Preço Médio": estatisticas_gasolina["Preço Médio"],
    "Gasolina - Desvio Padrão": estatisticas_gasolina["Desvio Padrão"],
    "Etanol - Preço Médio": estatisticas_etanol["Preço Médio"],
    "Etanol - Desvio Padrão": estatisticas_etanol["Desvio Padrão"]
})

analise_regional.sort_values(by=["Gasolina - Desvio Padrão", "Etanol - Desvio Padrão"], ascending=False, inplace=True)
analise_regional
```

```python
# Agrupando por região e calculando a média dos preços de gasolina e etanol
media_regiao = df_anp.groupby("ï»¿Regiao - Sigla").agg({
    "Valor de Venda": [
        lambda x: x[df_anp["Produto"] == "GASOLINA"].mean(),
        lambda x: x[df_anp["Produto"] == "ETANOL"].mean()
    ]
})

media_regiao.columns = ["Preço Médio Gasolina", "Preço Médio Etanol"]
media_regiao = media_regiao.reset_index().rename(columns={"ï»¿Regiao - Sigla": "Região"})
media_regiao = media_regiao.sort_values(by="Preço Médio Gasolina", ascending=False)
media_regiao
```
