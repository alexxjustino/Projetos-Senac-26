```python
import pandas as pd
import numpy as np
```

```python
dados = "https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv"

df = pd.read_csv(dados, sep=';', encoding='latin-1')
# Filtra apenas registros da capital e do ano de 2020
df_capital_2020 = df[(df['regiao'] == 'Capital') & (df['ano'] == 2020)]

# Agrupa por delegacia (cisp) e soma apreensões de menores
apreensao_menores_por_dp = df_capital_2020.groupby('cisp')['aaapai'].sum()

# Seleciona as delegacias no top 25% de apreensões
limite_25 = apreensao_menores_por_dp.quantile(0.75)
delegacias_top_25 = apreensao_menores_por_dp[apreensao_menores_por_dp >= limite_25]

display(delegacias_top_25)
```
